```python
import pandas as pd
import math
```

```python
df_protein = pd.read_csv("./protein_input.csv", index_col=0)
```

```python
df_interaction = pd.read_csv("./interaction_input.csv")
```

```python
df_complex = pd.read_csv("./complex_input.csv", index_col=0)
```

```python
df_gene = pd.read_csv("./gene_input.csv", index_col=1)
```

```python
lig_list = []
rec_list = []
type_list = []
for i in range(len(df_interaction)):
    name_a = df_interaction.iloc[i]['partner_a']
    name_b = df_interaction.iloc[i]['partner_b']
    if name_a in df_gene.index:
        prots_a = [name_a]
        receptor_a = df_protein.loc[name_a]['receptor'] + df_protein.loc[name_a]['integrin']
        secreted_a = df_protein.loc[name_a]['secreted'] + df_protein.loc[name_a]['secreted_highlight']
        transmem_a = df_protein.loc[name_a]['transmembrane']
    else:
        prots_a = list(df_complex.loc[name_a][['uniprot_1','uniprot_2','uniprot_3','uniprot_4']])
        prots_a = [g for g in prots_a if g==g]
        receptor_a = df_complex.loc[name_a]['receptor'] + df_complex.loc[name_a]['integrin']
        secreted_a = df_complex.loc[name_a]['secreted'] + df_complex.loc[name_a]['secreted_highlight']
        transmem_a = df_complex.loc[name_a]['transmembrane']
    if name_b in df_gene.index:
        prots_b = [name_b]
        receptor_b = df_protein.loc[name_b]['receptor'] + df_protein.loc[name_b]['integrin']
        secreted_b = df_protein.loc[name_b]['secreted'] + df_protein.loc[name_b]['secreted_highlight']
        transmem_b = df_protein.loc[name_b]['transmembrane']
    else:
        prots_b = list(df_complex.loc[name_b][['uniprot_1','uniprot_2','uniprot_3','uniprot_4']])
        prots_b = [g for g in prots_b if g==g]
        receptor_b = df_complex.loc[name_b]['receptor'] + df_complex.loc[name_b]['integrin']
        secreted_b = df_complex.loc[name_b]['secreted'] + df_complex.loc[name_b]['secreted_highlight']
        transmem_b = df_complex.loc[name_b]['transmembrane']

    genes_a = []
    for prot in prots_a:
        if len(df_gene.loc[prot].shape) > 1:
            genes_a.append(df_gene.loc[prot]['gene_name'][0])
        else:
            genes_a.append(df_gene.loc[prot]['gene_name'])
    genes_b = []
    for prot in prots_b:
        if len(df_gene.loc[prot].shape) > 1:
            genes_b.append(df_gene.loc[prot]['gene_name'][0])
        else:
            genes_b.append(df_gene.loc[prot]['gene_name'])

    if receptor_a and secreted_b:
        type_list.append('Secreted Signaling')
        lig_list.append(genes_b)
        rec_list.append(genes_a)
    elif receptor_b and secreted_a:
        type_list.append('Secreted Signaling')
        lig_list.append(genes_a)
        rec_list.append(genes_b)
    elif transmem_a and transmem_b:
        type_list.append('Cell-Cell Contact')
        lig_list.append(genes_a)
        rec_list.append(genes_b)
```

```python
lig_names = []
rec_names = []
for i in range(len(lig_list)):
    lig_names.append('_'.join(lig_list[i]))
    rec_names.append('_'.join(rec_list[i]))
```

```python
df_ligrec_human = pd.DataFrame(data={0:lig_names, 1: rec_names, 2: ['nan' for i in range(len(lig_list))], 3: type_list})
```

```python
df_ligrec_human
```

```python
df_hom = pd.read_csv("./HOM_AllOrganism.rpt.txt", sep='\t')

def human_to_mouse(df_hom, human_gene):
    df_tmp = df_hom[df_hom["Symbol"]==human_gene]
    geneid = df_tmp[df_tmp["Common Organism Name"]=="human"]["HomoloGene ID"].values
    if len(geneid) == 0:
        return None
    else:
        df_tmp = df_hom[df_hom["HomoloGene ID"]==geneid[0]]
        mouse_gene = df_tmp[df_tmp["Common Organism Name"]=="mouse, laboratory"]["Symbol"].values
        if len(mouse_gene) == 0:
            return None
        else:
            return mouse_gene[0]
```

```python
lig_list_mouse = []
rec_list_mouse = []
type_list_mouse = []
for i in range(len(lig_list)):
    ligs = lig_list[i]
    recs = rec_list[i]
    ligs_mouse = []
    recs_mouse = []
    incld = True
    for lig in ligs:
        lig_mouse = human_to_mouse(df_hom, lig)
        if lig_mouse is None: incld = False
        ligs_mouse.append(lig_mouse)
    for rec in recs:
        rec_mouse = human_to_mouse(df_hom, rec)
        if rec_mouse is None: incld = False
        recs_mouse.append(rec_mouse)
    if incld:
        lig_list_mouse.append(ligs_mouse)
        rec_list_mouse.append(recs_mouse)
        type_list_mouse.append(type_list[i])
```

```python
lig_names_mouse = []
rec_names_mouse = []
for i in range(len(lig_list_mouse)):
    lig_names_mouse.append('_'.join(lig_list_mouse[i]))
    rec_names_mouse.append('_'.join(rec_list_mouse[i]))
```

```python
df_ligrec_mouse = pd.DataFrame(data={0:lig_names_mouse, 1: rec_names_mouse, 2: ['nan' for i in range(len(lig_list_mouse))], 3: type_list_mouse})
```
